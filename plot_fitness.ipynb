```python
import os
import json
import matplotlib.pyplot as plt

FILENAME_PREFIX = "generation"
generation_data = []

# Loop over all generation files
i = 0
while True:
    filename = f"{FILENAME_PREFIX}{i}.json"
    if not os.path.exists(filename):
        break

    with open(filename, "r") as f:
        data = json.load(f)
        if len(data) == 0:
            i += 1
            continue
        # Get the best fitness for this generation
        best_fitness = max(ind["fitness"] for ind in data)
        generation_data.append((i, best_fitness))
    i += 1

# Plotting
generations = [gen for gen, _ in generation_data]
fitnesses = [fit for _, fit in generation_data]

plt.figure(figsize=(10, 5))
plt.plot(generations, fitnesses, marker='o')
plt.title("Best Fitness per Generation")
plt.xlabel("Generation")
plt.ylabel("Best Fitness")
plt.grid(True)
plt.xticks(range(0, len(generations), 1))
plt.tight_layout()
plt.show()
```
